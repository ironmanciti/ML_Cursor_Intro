- ë‹¨ìˆœ ì„ í˜•íšŒê·€ì˜ ê°œë…ê³¼ ì›ë¦¬ ì´í•´
- ë‹¨ë³€ìˆ˜, ë‹¤ë³€ìˆ˜, ì „ì²´ íŠ¹ì„±ì„ ì‚¬ìš©í•œ ì„ í˜•íšŒê·€ ëª¨ë¸ ë¹„êµ
- ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ ì§€í‘œ (RÂ², MSE) ì´í•´
- ë°ì´í„° ì‹œê°í™”ë¥¼ í†µí•œ íšŒê·€ ë¶„ì„ ê²°ê³¼ í•´ì„

## ë°ì´í„°ì…‹ ì†Œê°œ
**ìº˜ë¦¬í¬ë‹ˆì•„ ì£¼íƒê°€ê²© ë°ì´í„°ì…‹ (California Housing Dataset)**
- **ëª©í‘œ**: ìº˜ë¦¬í¬ë‹ˆì•„ ì§€ì—­ì˜ ì£¼íƒê°€ê²© ì˜ˆì¸¡
- **íŠ¹ì„± ìˆ˜**: 8ê°œ (MedInc, HouseAge, AveRooms, AveBedrms, Population, AveOccup, Latitude, Longitude)
- **ìƒ˜í”Œ ìˆ˜**: 20,640ê°œ
- **ëª©í‘œ ë³€ìˆ˜**: ì£¼íƒê°€ê²© (ë‹¨ìœ„: $100,000)

ì´ ë…¸íŠ¸ë¶ì—ì„œëŠ” ë‹¤ìŒê³¼ ê°™ì€ ë¶„ì„ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤:
1. **ë‹¨ë³€ìˆ˜ ì„ í˜•íšŒê·€**: MedInc(ì¤‘ê°„ ì†Œë“) â†’ ì£¼íƒê°€ê²©
2. **ë‹¤ë³€ìˆ˜ ì„ í˜•íšŒê·€**: MedInc + AveRooms â†’ ì£¼íƒê°€ê²©  
3. **ì „ì²´ íŠ¹ì„± ì„ í˜•íšŒê·€**: ëª¨ë“  8ê°œ íŠ¹ì„± â†’ ì£¼íƒê°€ê²©

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ import
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split  # ê°œì„ ëœ ë°ì´í„° ë¶„í• ì„ ìœ„í•´ ì¶”ê°€
from sklearn.preprocessing import MinMaxScaler  # ë°ì´í„° ì •ê·œí™”ë¥¼ ìœ„í•´ ì¶”ê°€
import pandas as pd

# Windowsì—ì„œ í•œê¸€ í°íŠ¸ ì„¤ì • (í•„ìš”ì‹œ ì£¼ì„ í•´ì œ)
import matplotlib.font_manager as fm
font_list = [font.name for font in fm.fontManager.ttflist]
if 'Malgun Gothic' in font_list:
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif 'NanumGothic' in font_list:
    plt.rcParams['font.family'] = 'NanumGothic'

plt.rcParams['axes.unicode_minus'] = False  # ë§ˆì´ë„ˆìŠ¤ ê¸°í˜¸ ê¹¨ì§ ë°©ì§€

# ì‹œê°í™” ì„¤ì •
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. ë°ì´í„° ë¡œë“œ ë° íƒìƒ‰

```python
# California Housing ë°ì´í„°ì…‹ ë¡œë“œ
# ì£¼íƒ ê°€ê²© ì˜ˆì¸¡ì„ ìœ„í•œ ë°ì´í„° (íŠ¹ì„± ê°„ ìƒê´€ê´€ê³„ê°€ ë‹¤ì–‘í•¨)
housing = datasets.fetch_california_housing()

print("íŠ¹ì„± ì´ë¦„:")
print(housing.feature_names)
print(f"\në°ì´í„° í˜•íƒœ: {housing.data.shape}")
print(f"ëª©í‘œ ë³€ìˆ˜ í˜•íƒœ: {housing.target.shape}")
print(f"\níŠ¹ì„± ì„¤ëª…:")
for i, name in enumerate(housing.feature_names):
    print(f"  {i}: {name}")
```

### ë°ì´í„°ì…‹ ê¸°ë³¸ ì •ë³´

```python
# DataFrameìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ë°ì´í„° íƒìƒ‰
df_housing = pd.DataFrame(data=housing.data, columns=housing.feature_names)
df_housing['target'] = housing.target  # ëª©í‘œ ë³€ìˆ˜ë„ ì¶”ê°€

print("DataFrame í˜•íƒœ:")
print(df_housing.head())
print(f"\nDataFrame ì •ë³´:")
print(df_housing.info())
```

### ë°ì´í„°ì…‹ ìƒì„¸ ì •ë³´

```python
# MedInc(ì¤‘ê°„ ì†Œë“) íŠ¹ì„±ë§Œ ì‚¬ìš©í•˜ì—¬ ë‹¨ë³€ìˆ˜ ì„ í˜•íšŒê·€ ë°ì´í„° ì¤€ë¹„
# reshape(-1, 1): 1ì°¨ì› ë°°ì—´ì„ 2ì°¨ì›ìœ¼ë¡œ ë³€í™˜ (sklearn ìš”êµ¬ì‚¬í•­)
X_medinc = df_housing['MedInc'].values.reshape(-1, 1)
y = housing.target

print(f"MedInc ë°ì´í„° í˜•íƒœ: {X_medinc.shape}")
print(f"ëª©í‘œ ë³€ìˆ˜ í˜•íƒœ: {y.shape}")
print(f"MedInc ë²”ìœ„: ${X_medinc.min():.2f} ~ ${X_medinc.max():.2f}")
print(f"MedInc í‰ê· : ${X_medinc.mean():.2f}")
```

### ë°ì´í„° ë¶„í• 

```python
# ê°œì„ ëœ ë°ì´í„° ë¶„í•  ë°©ë²• ì‚¬ìš©
# train_test_splitì„ ì‚¬ìš©í•˜ì—¬ ë” ì²´ê³„ì ìœ¼ë¡œ ë°ì´í„° ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(
    X_medinc, y, test_size=0.2, random_state=42
)

# MinMaxScalerë¥¼ ì‚¬ìš©í•œ ë°ì´í„° ì •ê·œí™”
scaler_single = MinMaxScaler()
X_train_scaled = scaler_single.fit_transform(X_train)
X_test_scaled = scaler_single.transform(X_test)

print(f"í›ˆë ¨ ë°ì´í„° í¬ê¸°: {X_train.shape}")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„° í¬ê¸°: {X_test.shape}")
print(f"ì •ê·œí™” ì „ MedInc ë²”ìœ„: {X_train.min():.2f} ~ {X_train.max():.2f}")
print(f"ì •ê·œí™” í›„ MedInc ë²”ìœ„: {X_train_scaled.min():.2f} ~ {X_train_scaled.max():.2f}")
```

```python
# ì„ í˜•íšŒê·€ ëª¨ë¸ ìƒì„± ë° í›ˆë ¨ (ì •ê·œí™”ëœ ë°ì´í„° ì‚¬ìš©)
regr = linear_model.LinearRegression()

# ì •ê·œí™”ëœ í›ˆë ¨ ë°ì´í„°ë¡œ ëª¨ë¸ í•™ìŠµ
regr.fit(X_train_scaled, y_train)

print("ëª¨ë¸ í›ˆë ¨ ì™„ë£Œ!")
print(f"íšŒê·€ ê³„ìˆ˜ (ê¸°ìš¸ê¸°): {regr.coef_[0]:.2f}")
print(f"ì ˆí¸: {regr.intercept_:.2f}")
print(f"\níšŒê·€ì‹: ì£¼íƒê°€ê²© = {regr.coef_[0]:.2f} * MedInc(ì •ê·œí™”) + {regr.intercept_:.2f}")
print(f"í•´ì„: ì •ê·œí™”ëœ ì¤‘ê°„ ì†Œë“ì´ 1ë‹¨ìœ„ ì¦ê°€í•˜ë©´ ì£¼íƒê°€ê²©ì´ {regr.coef_[0]:.2f}ë§Œ ë‹¬ëŸ¬ ì¦ê°€")
```

```python
# ëª¨ë¸ ì˜ˆì¸¡ ë° ì„±ëŠ¥ í‰ê°€ (ì •ê·œí™”ëœ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì‚¬ìš©)
y_pred = regr.predict(X_test_scaled)

# ëª¨ë¸ ì„±ëŠ¥ í‰ê°€
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("="*50)
print("ë‹¨ë³€ìˆ˜ ì„ í˜•íšŒê·€ ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ (ì •ê·œí™” ì ìš©)")
print("="*50)
print(f"í‰ê·  ì œê³± ì˜¤ì°¨ (MSE): {mse:.2f}")
print(f"ê²°ì •ê³„ìˆ˜ (RÂ²): {r2:.2f}")
print(f"\ní•´ì„:")
print(f"- RÂ² = {r2:.2f}ëŠ” MedIncê°€ ì£¼íƒê°€ê²©ì˜ {r2*100:.0f}%ë¥¼ ì„¤ëª…í•¨")
print(f"- ë‚˜ë¨¸ì§€ {(1-r2)*100:.0f}%ëŠ” ë‹¤ë¥¸ ìš”ì¸ë“¤ì— ì˜í•´ ê²°ì •ë¨")
print(f"- MSE = {mse:.2f}ëŠ” ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ì˜ í‰ê·  ì œê³± ì°¨ì´")
print(f"- RMSE = {np.sqrt(mse):.2f} (í‰ê·  ì˜ˆì¸¡ ì˜¤ì°¨)")
```

## ì „ì²´ íŠ¹ì„± ì„ í˜•íšŒê·€ (ëª¨ë“  íŠ¹ì„± â†’ ì£¼íƒê°€ê²©)

### ì „ì²´ íŠ¹ì„± ë°ì´í„° ë¶„í• 

```python
# ì „ì²´ íŠ¹ì„±ì„ ì‚¬ìš©í•œ ì„ í˜•íšŒê·€ ë°ì´í„° ì¤€ë¹„
X_all = df_housing.drop('target', axis=1).values  # target ì»¬ëŸ¼ ì œì™¸í•œ ëª¨ë“  íŠ¹ì„±
y = housing.target

print(f"ì „ì²´ íŠ¹ì„± ë°ì´í„° í˜•íƒœ: {X_all.shape}")
print(f"ì‚¬ìš©ëœ íŠ¹ì„±: {list(df_housing.columns[:-1])}")  # target ì œì™¸
print(f"íŠ¹ì„± ìˆ˜: {X_all.shape[1]}ê°œ")
```

```python
# ì „ì²´ íŠ¹ì„± ì„ í˜•íšŒê·€ë¥¼ ìœ„í•œ ë°ì´í„° ë¶„í• 
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(
    X_all, y, test_size=0.2, random_state=42
)

# MinMaxScalerë¥¼ ì‚¬ìš©í•œ ì „ì²´ íŠ¹ì„± ë°ì´í„° ì •ê·œí™”
scaler_all = MinMaxScaler()
X_train_all_scaled = scaler_all.fit_transform(X_train_all)
X_test_all_scaled = scaler_all.transform(X_test_all)

print(f"í›ˆë ¨ ë°ì´í„° í¬ê¸°: {X_train_all.shape}, {y_train_all.shape}")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„° í¬ê¸°: {X_test_all.shape}, {y_test_all.shape}")
print(f"ì •ê·œí™” ì „ ë²”ìœ„: {X_train_all.min():.2f} ~ {X_train_all.max():.2f}")
print(f"ì •ê·œí™” í›„ ë²”ìœ„: {X_train_all_scaled.min():.2f} ~ {X_train_all_scaled.max():.2f}")
```

```python
# ì „ì²´ íŠ¹ì„± ì„ í˜•íšŒê·€ ëª¨ë¸ í›ˆë ¨ (ì •ê·œí™”ëœ ë°ì´í„° ì‚¬ìš©)
regr_all = linear_model.LinearRegression()
regr_all.fit(X_train_all_scaled, y_train_all)

print("ì „ì²´ íŠ¹ì„± ì„ í˜•íšŒê·€ ëª¨ë¸ í›ˆë ¨ ì™„ë£Œ!")
print(f"íšŒê·€ê³„ìˆ˜: {regr_all.coef_}")
print(f"ì ˆí¸: {regr_all.intercept_:.2f}")
print(f"ì‚¬ìš©ëœ íŠ¹ì„± ìˆ˜: {len(regr_all.coef_)}ê°œ (ì •ê·œí™” ì ìš©)")
print(f"íŠ¹ì„±ëª…: {list(df_housing.columns[:-1])}")  # target ì œì™¸
```

```python
# ì „ì²´ íŠ¹ì„± ëª¨ë¸ ì˜ˆì¸¡ ë° ì„±ëŠ¥ í‰ê°€ (ì •ê·œí™”ëœ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì‚¬ìš©)
y_pred_all = regr_all.predict(X_test_all_scaled)

mse_all = mean_squared_error(y_test_all, y_pred_all)
r2_all = r2_score(y_test_all, y_pred_all)

print("="*50)
print("ì „ì²´ íŠ¹ì„± ì„ í˜•íšŒê·€ ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ (ì •ê·œí™” ì ìš©)")
print("="*50)
print(f"í‰ê·  ì œê³± ì˜¤ì°¨ (MSE): {mse_all:.2f}")
print(f"ê²°ì •ê³„ìˆ˜ (RÂ²): {r2_all:.2f}")
print(f"\ní•´ì„:")
print(f"- RÂ² = {r2_all:.2f}ëŠ” ëª¨ë“  íŠ¹ì„±ì´ ì£¼íƒê°€ê²©ì˜ {r2_all*100:.0f}%ë¥¼ ì„¤ëª…í•¨")
print(f"- ë‹¨ë³€ìˆ˜ ëª¨ë¸ ëŒ€ë¹„ ì„±ëŠ¥ í–¥ìƒ: RÂ² {r2:.2f} â†’ {r2_all:.2f} ({r2_all-r2:+.2f})")
```

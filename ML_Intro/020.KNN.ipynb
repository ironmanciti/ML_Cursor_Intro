# 020.KNN (K-Nearest Neighbors, 최근접 이웃)

### iris dataset

iris.feature_names : 

Sepal Length : 꽃받침 길이  
Sepal Width  : 꽃받침 너비   
Petal Length  : 꽃잎 길이  
Petal Width   :  꽃잎 너비 

Species (꽃의 종류) :  setosa / versicolor / virginica 의 3종류로 구분된다.

**꽃받침 길이, 너비 두가지 feature 를 가지고 KNN 알고리즘을 이용하여 꽃의 종류 분류**

**neighbors.KNeighborsClassifier(n_neighbors, weights=weights)**

- weights :  

    - uniform : uniform weights. 모든 neighbor 의 가중치를 동일하게 취급  
    - distance : neighbor 의 거리에 반비례하여 가중치 조정

```python
# 필요한 라이브러리 import
import numpy as np
from sklearn import neighbors, datasets
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows용 한글 폰트
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 폰트 설정 확인
print("현재 설정된 폰트:", plt.rcParams['font.family'])
```

```python
# iris 데이터셋 로드 및 기본 정보 확인
iris = datasets.load_iris()

print("데이터 형태:", iris.data.shape)
print("특성 이름:", iris.feature_names)
print("클래스 이름:", iris.target_names)
```

```python
# 꽃받침 길이, 너비 두가지 feature 선택 (2D 시각화를 위해)
X = iris.data[:, :2]  # 첫 번째, 두 번째 특성 (꽃받침 길이, 너비)
y = iris.target       # 타겟 변수 (꽃의 종류)

print("특성 데이터 (꽃받침 길이, 너비):")
print(X[:5])
print()
print("타겟 데이터 (꽃의 종류):")
print(y[:5])
```

```python
# 훈련 데이터와 테스트 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"훈련 데이터 크기: {X_train.shape}")
print(f"테스트 데이터 크기: {X_test.shape}")
```

### KNN 모델 생성 및 훈련

```python
# KNN 분류기 생성 (k=15, 균등 가중치)
clf = neighbors.KNeighborsClassifier(n_neighbors=15, weights='uniform')
clf.fit(X_train, y_train)  # 모델 훈련
```

```python
# 테스트 데이터에 대한 예측 수행
y_predict = clf.predict(X_test)
print("예측 결과:")
print(y_predict)
```

```python
# 혼동 행렬 계산
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predict)
print("혼동 행렬:")
print(cm)
print()
print("혼동 행렬 해석:")
print("- 행: 실제 클래스, 열: 예측 클래스")
print("- 대각선: 올바르게 분류된 샘플 수")
print("- 비대각선: 잘못 분류된 샘플 수")
```

### 모델 성능 평가

```python
# 정확도 계산
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)
print(f"모델 정확도: {accuracy:.4f} ({accuracy*100:.2f}%)")
```

### 결정 경계 시각화

```python
# 결정 경계 시각화 함수
def plot_decision_boundary(X, y, classifier, title):
    # 색상 맵 설정
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
    
    # 그리드 생성
    h = 0.02  # 그리드 간격
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    # 그리드 포인트에 대한 예측
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # 그래프 그리기
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolors='black', s=50)
    
    plt.xlabel('꽃받침 길이 (cm)')
    plt.ylabel('꽃받침 너비 (cm)')
    plt.title(title)
    plt.colorbar()
    plt.show()

# 결정 경계 시각화
plot_decision_boundary(X_train, y_train, clf, 'KNN 결정 경계 (k=15)')
```

```python
plot_decision_boundary(X_test, y_test, clf, 'KNN 결정 경계 (k=15)')
```
